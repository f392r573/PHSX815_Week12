```python
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.read_excel ('IPEDS_data.xlsx')

data1 = pd.DataFrame(dataframe,columns=["Graduation rate - Bachelor degree within 4 years, total", "Tuition and fees, 2013-14"])

plt.figure()

plt.scatter(data1["Graduation rate - Bachelor degree within 4 years, total"],data1["Tuition and fees, 2013-14"])
plt.show()
```

I do the relationship between Graduation rate - Bachelor degree within 4 years, total and Tuition and fees, 2013-14. It looks like there is some correlation between these.

```python
data2 = pd.DataFrame(dataframe,columns=["Applicants total","Offers Bachelor's degree","Offers Master's degree","Offers Doctor's degree - research/scholarship"])
data2=data2.dropna()

data2["Offers Bachelor's degree"] =data2["Offers Bachelor's degree"].astype('category').cat.codes
data2["Offers Master's degreee"] =data2["Offers Master's degree"].astype('category').cat.codes
data2["Offers Doctor's degree - research/scholarship"] =data2["Offers Doctor's degree - research/scholarship"].astype('category').cat.codes
plt.figure()
plt.scatter(data2["Applicants total"],data2["Offers Bachelor's degree"],data2["Offers Master's degreee"],data2["Offers Doctor's degree - research/scholarship"])
plt.show()
```

I do the relationship between Applicants total, offers Bachelor's degree, Offers Master's degree and Offers Doctor's degree. It looks like there isn't any correlation between these

```python
data3 = pd.DataFrame(dataframe,columns=["Graduation rate - Bachelor degree within 4 years, total","Control of institution"])

data3 = data3.dropna()
data3 = data3.groupby(["Control of institution"]).sum()

print(data3)
plt.figure()
data3.plot.bar()
plt.show()

```

I do the relationship between Graduation rate - Bachelor degree within 4 years, total and Control of institution. It shows there is  big correlation between these.
